# Variance 

## What is variance? 
	
The variance is defined as the average of the square of the difference of each number in the data to the mean of the data. It is denoted as $\sigma^2$.

__It is a measure that refers to the spread of numbers in a data set.__ A large variance means that the numbers in a set are far from the mean and each other. A small variance means that the numbers are close to the mean and to each other in value. 

> Another measure of the spread of numbers in a data set is the mean absolute deviation, which you can read about here: [link](https://statistics.laerd.com/statistical-guides/measures-of-spread-absolute-deviation-variance.php)

We have seen $\sigma$ as the standard deviation of a given dataset: __Variance is simply the square of standard deviation (We could also say standard deviation is the square root of variance)__. 

## How to compute variance

The variance of a sample of n numbers (for example, the weight of 100 chickens) is calculated by:

1. Taking the differences between each element of the sample and the sample mean, where the sample mean is denoted by $\mu$, 
2. Summing together all the squares of the differences 
3. Dividing the sum by n - 1

$$ \large \sigma^2 = \frac{\sum_{i=1}^{n}(x_i-\mu)^2}{n-1}$$

Here, $x_i$ refers to the ith element of the sample of n numbers. 
- - -
Let's see this in action: 

Say I had a sample of the masses of 10 books in my bookcase, measured in grams.

```python
import random

def get_variance(mass_of_books, print_=True):
    
    size_of_sample = len(mass_of_books)
    
    mean_mass_of_books = sum(mass_of_books)/size_of_sample
    
    differences = [i - mean_mass_of_books for i in mass_of_books]
    
    square_of_differences = [i**2 for i in differences]
    
    sum_of_square_of_differences = sum(square_of_differences)
    
    variance = (1/(size_of_sample - 1))*sum_of_square_of_differences
    
    if print_:
        print("The variance of the mass of a sample of 10 books in my bookcase is:", round(variance,1), "grams squared.")
    
    return variance

mass_of_books = [random.randint(100, 300) for i in range(10)]

print("The mass of a sample of 10 books in my bookcase is:", mass_of_books)

_= get_variance(mass_of_books)
```

You may ask yourself why we square the differences when computing variance. If we didn't square the difference of each element in a data set to the mean value of the data, positive and negative differences would cancel each other out when computing the average difference of all the elements in the data set to the mean.

You may also ask yourself why we divide by $(n-1)$ instead of $n$: 
* If we are computing the variance of a **population** of size n, we divide by n. 
* When computing the variance of a **sample** of numbers of size n, we divide by n-1. 

## Intepretation of variance 

* A high variance indicates that the data points are very spread out from the mean, and from one another. 

* A low variance indicates that the data points are very narrowly spread about the mean, and are close to each other. 

* A variance of zero indicates that all of the data values are identical.

# Covariance

## What is covariance?

Covariance provides an insight into how two variables are __related__ to one another. It refers to the measure of how two random variables in a data set will __change together__.

Suppose we collect a sample of data of the age and heights of 100 people. Is there a relationship between the height and age of the people in our sample? Are older people taller, and younger people shorter? 


## How to compute covariance

The covariance of sample of size n of two random variables (for example, the age and heights of 100 people) is calculated by:

1. For each variable, taking the differences between each element of the sample and respective variable sample mean,
2. Computing the sum of the product of the differences
3. Dividing this sum by n - 1

Mathematically, the covariance between two samples of size n of random variables X and Y is denoted as $\sigma_{XY}$, as is defined as follows: 

$$ \large \sigma_{XY} = \frac{\sum_{i=1}^{n}(x_i -\mu_x)(y_i - \mu_y)}{n-1}$$ 

where $X$ and $Y$ are the two samples of random variables of n elements each. 

We want to measure how values in $X$ change with observed changes in $Y$ values, and vice versa. 

In this equation, 
* $x_i$  is the ith element of variable $X$
* $y_i$ = ith element of variable $Y$
* $n$ = number of data points, where __$n$ must be same for $X$ and $Y$__
* $\mu_x$ = mean of $X$
* $\mu_y$ = mean of $Y$

Notice that if Y = X, then the equation for the covariance is the same as that of the variance. 
> In other words, the variance of a variable is the covariance of the variable with itself. 

## Intepretation of covariance:

* A positive covariance indicates that higher than average values of one variable tend to pair with higher than average values of the other variable, and lower than average values of one variable tend to pair with lower than average values of the other variable.

* Negative covariance indicates that lower than average values of one variable tend to pair with higher than average values of the other variable, and higher than average values of one variable tend to pair with lower than average values of the other variable. 

* If the two random variables are independent from each other,the covariance will be zero. However, __a covariance of zero does not necessarily mean that the variables are independent__. A nonlinear relationship can exist that still would result in a covariance value of zero. 

This behavior can be further explained using the scatter plots below: 

<img src="images/covariance.gif" width=500>

A large negative covariance value show an inverse relationship between values at x and y axes. i.e. y decreases as x increases. This is shown by the scatter plot on the left. The middle scatter plot shows values spread all over the plot, reflecting the fact that variables on x and y axes can not be related in terms of how they vary together. The covariance value for such variables would be very close to zero. In the scatter plot on right, we see a strong relationship between values at x and y axes i.e. y increases as x increases. 


## Example: 

Let's follow up with the example used in the previous section. 

I want to know whether there is a relationship between the mass of the sample of 10 books from my bookcase and the number of pages contained in each of the 10 books.

```python
n_pages_in_books = [random.randint(100, 450) for i in range(10)]
print("The number of pages in the books of a sample of 10 books in my bookcase is:", n_pages_in_books)
```

First, let's plot a scatterplot of the mass of the sample of books against the number of pages in the books: 

```python
import matplotlib.pyplot as plt 
%matplotlib inline 

fig = plt.figure(figsize=(6, 6))
plt.scatter(mass_of_books, n_pages_in_books)
plt.xlabel('mass (grams)')
plt.ylabel('number of pages');
```

Next, let's compute the covariance: 

```python
def get_covariance(mass_of_books, n_pages_in_books, print_=True):
    
    size_of_sample = len(mass_of_books)

    mean_mass_of_books = sum(mass_of_books)/size_of_sample
    
    mean_n_pages_in_books = sum(n_pages_in_books)/size_of_sample

    products = [(i-mean_mass_of_books)*(j-mean_n_pages_in_books) for i, j in zip(mass_of_books, n_pages_in_books)]
    
    sum_of_products = sum(products)
    
    covariance = (1/(size_of_sample - 1)*sum_of_products)
    
    if print_:
        print("The covariance of the mass of a sample of 10 books in my bookcase and the number of pages in the books is:", 
             round(covariance, 1), " 'gram-pages' ")
    
    return covariance

_ = get_covariance(mass_of_books, n_pages_in_books)
```

>__Covariance is not standardized and has the units of the quantities one is computing the covariance of.  Also, covariance values can range from negative infinity to positive infinity.__

# Correlation

## What is correlation?	

We saw how the covariance can identify the degree to which two random variables tend to vary together. 

In data science practice, we typically to look at the correlation rather than the covariance of two random variables because correlation does not depend on the scale of either random variable involved and is more interpretable as a result. We need to normalize this degree of variation into a standard unit, with results independent of the units of the data. We achieve this with a derived normalized measure, called _correlation_. 

The __correlation__ between two random variables X and Y is defined as the covariance of X and Y divided by the square root of the product of the standard deviations of $X$ and $Y$. This _normalization_ sets the range of the correlation to lie between the values of -1 to 1. 

Mathematically, the correlation between two random variables X and Y is calculated as:

$$correlation(X,Y) = \frac{\sigma_{XY}}{\sqrt{\sigma_X\sigma_Y}}$$

where $\sigma_{X}$ and $\sigma_{Y}$ are, respectively, the standard deviations of X and Y. 

## How to compute correlation

In statistics, we measure different types of correlations for detailed relationship analysis: 

* Pearson's r correlation
    * The Pearson correlation measures the extent to which two variables are __linearly__ related
* Spearman rank correlation
    * The Spearman rank correlation assesses how well the relationship between two variables can be described using a monotonic function 
* Kendall rank correlation 
    * The Kendall rank correlation measures the ordinal association between two measured quantities
* Point-Biserial correlation
    * The point-biserial correlation is used when one of the two variables is dichotomous (i.e. has only two possible outcomes)

For now, we will focus on the Pearson's r correlation as it is the most widely used correlation statistic to measure the degree of the relationship between two __linearly__ related variables. 


### Calculating Pearson's r correlation

Pearson's correlation coefficient when applied to a sample is represented by _r_ and is referred to as the sample Pearson correlation coefficient. 

The Pearson correlation coefficient of sample of size n of two random variables (for example, the weight and height of 100 chickens) is calculated by:

1. For each variable, taking the differences between each element of the sample and respective variable sample mean,
2. Computing the sum of the product of the differences
3. Dividing this sum by the product of the standard deviation of X and Y, $\sigma_X$ and $\sigma_Y$, respectively. 

Mathematically, given paired data $X$ and $Y$ of n elements each, the sample Pearson correlation coefficient is defined as: 

$$ r_{XY} = \frac{\sum_{i=1}^{n}(x_i -\mu_x)(y_i - \mu_y)} {\sqrt{(\sum_{i=1}^{n}x_i - \mu_x)^2 (\sum_{i=1}^{n}y_i-\mu_y)^2}}$$

Just like in the case of covariance, 

* $x_i$ = ith element of variable $X$

* $y_i$ = ith element of variable $Y$

* $n$ = number of data points (__$n$ must be same for $X$ and $Y$__)

* $\mu_x$ = sample mean of $X$

* $\mu_y$ = sample mean of $Y$

Notice that the Pearson correlation coefficient is symmetric under the exchange of X and Y, that is, $Correlation(X,Y) = Correlation(Y,X)$. 

## Interpreting Pearson's r correlation values

> __The correlation formula shown above will always result in values in a range between -1 and 1__

* A value of 1 implies that a linear equation describes the relationship between X and Y, with all data points lying on a line for which X and Y change together in the same direction. 

* A value of -1 implies all data points lie on a line for which X and Y change together in opposite directions. 

* A value of 0 implies there is no __linear__ correlation between the variables. 

- - - 
The Pearson correlation coefficient of two variables measures whether there is a linear correlation between the two variables. In the plot below, we can see the result of computing the Pearson correlation coefficient for different data: 

![](images/correlation.png)

__Note that a value of zero does not imply there is no relation between the variables, just that there is not a _linear_ relationship between them.__

- - - 

For the sake of completeness and to show an example computation of correlation, let's compute the Pearson correlation of the mass and number of pages of the sample of 10 books from my bookcase: 

```python
import math

def get_pearson_correlation(mass_of_books, n_pages_in_books, print_=True):
    variance_mass_books = get_variance(mass_of_books, print_=False)
    variance_n_pages_books = get_variance(n_pages_in_books, print_=False)
    covariance_mass_and_pages = get_covariance(mass_of_books, n_pages_in_books, print_=False)
    
    pearson_r = covariance_mass_and_pages/math.sqrt(variance_mass_books*variance_n_pages_books)

    if print_:
        print("The Pearson r coefficient is:", round(pearson_r, 1))
    return 

pearson_r = get_pearson_correlation(mass_of_books,n_pages_in_books) 
```

## Additional example use cases

#### The Stock Market

Think about stock markets in terms of correlation. Stock market indexes tend to move together in similar directions. When the DOW Jones loses 5%, the S&P 500 usually loses around 5%. When the DOW Jones gains 5%, the S&P 500 usually gains around 5% because they are **highly correlated**.

On the other hand, there could also be negative correlation where you might observe that as the DOW Jones loses 5% of it value, Gold might gain 5%. Alternatively, if the Dow Jones gains 5% of its value, Gold may lose 5% of its value. That's a **negative correlation**. 

#### Social Media and Websites

Digital publishers want to maximize their understanding of the potential relationship between social media activity and visits to their website. For example, the digital publisher runs the correlation report between hourly Twitter mentions and visits for a two week period. The correlation is found to be r = 0.28, which indicates a medium, positive relationship between Twitter mentions and website visits.

#### Optimization for E-retailers
E-retailers are interested in driving increased revenue. For example, an e-retailer wants to compare a number of secondary success events (e.g., file downloads, product detail page views, internal search click-throughs, etc.) with weekly web revenue. They can quickly identify internal search click-throughs as having the highest correlation, which may indicate an area for optimization.
